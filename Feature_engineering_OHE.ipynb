#### Onehotencoding with many categories in a feature


```python
df_1=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/mercedes.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
```

```python
df_1.head()
```

```python
list_10 =df_1.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
list_10 .head()
```

```python
type(list_10)
```

```python
#converting to list
lst_10=list(list_10)
```

```python
lst_10
```

```python
for categories in lst_10:
    df_1[categories]=np.where(df_1['X1']==categories,1,0)
```

```python
lst_10.append('X1')
```
