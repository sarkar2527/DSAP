Bagging ensemble method

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag = BaggingClassifier(base_estimator = dt, max_samples = 0.5)
bag.fit(X_train,Y_train)
Y_pred_bag = bag.predict(X_test)
```

```python
acc_bag = accuracy_score(Y_test,Y_pred_bag)
print(acc_bag)
cm_bag = confusion_matrix(Y_test,Y_pred_bag)
print(cm_bag)
cr_bag = classification_report(Y_test,Y_pred_bag)
print(cr_bag)
```

```python
bag_score = []
ms = np.arange(0.1,1,0.1)
for i in ms:
    bag = BaggingClassifier(base_estimator = dt, max_samples = i)
    bag.fit(X_train,Y_train)
    Y_pred_bag = bag.predict(X_test)
    bag_score.append(accuracy_score(Y_test,Y_pred_bag))
    
print(bag_score)
    
```

```python
plt.plot(ms,bag_score,label = "accuracy")
plt.legend()
```

From the above the graph the best accuracy is attained at 0.9 max_samples but the reasonable accuracy is attained at 0.5 samples where the accuracy is 0.89 and f1-score of 0.89 for both the classes. Hence choosing 0.5 max_samples as best option in bagging which is not overfitting as well.
