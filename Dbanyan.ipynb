### 5. No of days difference between each purchase made by customers?

```python
df1['Date'] = pd.to_datetime(df1.day, format='%d')
df1 = df1.sort_values('day')

def AVG_diff(x):
    # Integer number of Days
    return x.diff().mean().total_seconds()/86400

df1.groupby('customer_id').Date.agg(['ordered_item_quantity', AVG_diff])
```
