```python
product_type_analysis=df1.groupby('product_type').agg({'orders':sum,'total_sales':sum}).reset_index()
product_type_analysis
```

### 4.Does the most sold product and least sold product has equal sales in every month or not?

```python
plt.figure(figsize=(35, 10))
plt.title('Total Monthly transaction by product type')
sns.countplot(df1['product_type'], hue = df1.month)
```

From the above graph we can say that most sold product(Sweets and Snacks) doesn't have equal sales in every month. It is varying but we can say almost equal.  Least product(Pooja_items) sold product has slowly decreasing in every month.

**5.Has the most sold product and least sold product has equal sales in every year or not?**
